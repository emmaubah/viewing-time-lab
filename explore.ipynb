```python

# TODO: import all the packages that you will need for this notebook here
from cleaner import clean_data
import statistics as stats
import matplotlib.pyplot as plt
```

```python
control_data = []
treatment_data = []

control_obj, treatment_obj = open("data/control.txt"), open("data/treatment.txt")
# read_control, read_treatment = control_obj.readlines(), treatment_obj.readlines()

for num1 in control_obj:
    control_data.append(num1.strip())
for num2 in treatment_obj:
    treatment_data.append(num2.strip())

control_obj.close()
treatment_obj.close()

# TODO: clean both list objects that you created above using the 'clean_data' function
cleaned_control = clean_data(control_data)
cleaned_treatment = clean_data(treatment_data)

print(control_data)
print(treatment_data)
print(cleaned_control)
print(cleaned_treatment)
```

## 2. Explore your Datasets

Now that you have your list objects created and cleaned, let's explore the descriptive statistics of these two files and make subsequent observations on our data.

```python
# TODO: Calculate the standard deviation of both the control and treatment group
control_stdev = stats.stdev(cleaned_control)
treatment_stdev = stats.stdev(cleaned_treatment)

# TODO: print your findings
print("The control group standard dev. is:", round(control_stdev, 2))
print("The treatment group standard dev. is: ", round(treatment_stdev, 2))
```

```python
# TODO: Calculate the mean of both the control and treatment group
control_mean = stats.mean(cleaned_control)
treatment_mean = stats.mean(cleaned_treatment)

# TODO: print your findings
print("The control group mean is:", round(control_mean, 2))
print("The treatment group mean is: ", round(treatment_mean, 2))
```

```python
# TODO: Are there any outliers in your control group? (use 2 * stdev to find these outliers)
outliers_control = []
for x1 in cleaned_control:
    if x1 > control_mean + 2*control_stdev:
        outliers_control.append(x1)
print(outliers_control)
```

```python
# TODO: Are there any outliers in your treatment group?
outliers_treatment = []
for x2 in cleaned_treatment:
    if x2 > treatment_mean + 2*treatment_stdev:
        outliers_treatment.append(x2)

print(outliers_treatment)
```
